```python
#Lets consider a pie chart.

#4 Parties
#labels of the party
party_name = ['Party A','Party B','Party C','PartyD']
#Votes received by the party
votes = [175000,50000,25000,50000]
#If we want to show one as standing out.
explode=(0,0,0,0)

fig1,ax1 = plt.pyplot.subplots()
ax1.pie(votes,explode=explode,labels=party_name,autopct='%1.1f%%',shadow=True,startangle=90)
ax1.axis('equal')

plt.pyplot.show()
        
```
